```python
import pandas as pd
df1 = pd.read_csv('data/train.csv', index_col=0)
df1.head()
```

```python
## 데이터 불러오기
df2 = pd.read_csv('data/time_series_data.csv')
df2.dropna(inplace=True)

## 변수명 변경
df2.columns = ['Period', 'Revenue', 'Sales_quantity', 'Average_cost',
               'payroll_region_mean']

## Period 변수 datetime으로 전환
df2.Period = pd.to_datetime(df2.Period)

## 데이터 확인
df2.head()
```

```python
### 선그래프
# 사이즈 변환
# import matplotlib.pyplot as plt
df2.plot(x='Period', y='Revenue', figsize=(10,6))
# plt.title('time series')
# plt.xlabel('time')
# plt.ylabel('revenue')
# plt.show()
```

```python
### 선그래프
# 사이즈 변환
import matplotlib.pyplot as plt
df2.plot(x='Period', y='Revenue', figsize=(10,6))
plt.title('time series')
plt.xlabel('time')
plt.ylabel('revenue')
plt.show()
```

```python
### 선그래프 y축에 2개 그리기
df2.Sales_quantity = df2.Sales_quantity*(1/10)
df2.plot(x='Period', y=['Average_cost','Sales_quantity'], figsize=(10,6))
plt.title('time series')
plt.xlabel('time')
plt.ylabel('Average_cost, Sales_quantity')
plt.show()
```

```python
df2.head()
```

```python
## .plot() 안에서 xlabel과 ylabel 표시하기
df2.plot(x = 'Period', y=['Revenue','Sales_quantity'], xlabel = 'time', ylabel='av_cost', figsize = (10,7))
```

```python
## 시간을 나타내는 변수가 index가 있다면 secondary_y 옵션을 사용하여 하나의 그래프에 y추가 가능
df3 = df2.set_index('Period').copy()

df2.plot(x='Period', y='Average_cost', figsize=(10,6))
df2['Revenue'].plot(secondary_y=True, style='g')
```

```python
## 시간을 나타내는 변수가 index가 있다면 secondary_y 옵션을 사용하여 하나의 그래프에 y추가 가능
df3 = df2.set_index('Period').copy()

df3['Sales_quantity'].plot(figsize=(10,7))
df3['Revenue'].plot(secondary_y=True, style='m')
```
